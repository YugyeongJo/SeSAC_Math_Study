# 함수 그래프 시각화

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-4, 4, 100)
print(x)
```

```python
y = x

plt.plot(x, y, label = "Y = X")
plt.legend()
plt.grid(linestyle = ":")
plt.show()
```

```python
y2 = x**2

plt.plot(x, y2, label = "Y = X^2")
plt.legend()
plt.grid(linestyle = ":")
plt.show()
```

```python
y3 = x**3

plt.plot(x, y3, label = "Y = X^3")
plt.legend()
plt.grid(linestyle = ":")
plt.show()
```

```python
def polynomial(x, n):
    y = x**n
    return y

x = np.linspace(-4, 4, 100)
y = polynomial(x, 1)
y2 = polynomial(x, 2)
y3 = polynomial(x, 3)

plt.plot(x, y, label = "Y = X")
plt.plot(x, y2, label = "Y = X^2")
plt.plot(x, y3, label = "Y = X^3")
plt.plot(x, polynomial(x, 4), label="y = x^4")

plt.title("Polynomial Curve")
plt.xlabel("X")
plt.ylabel("f(x)")
plt.legend()
plt.ylim(-2, 2)
plt.xlim(-2, 2)

plt.grid(linestyle = ":")
plt.show()
```
