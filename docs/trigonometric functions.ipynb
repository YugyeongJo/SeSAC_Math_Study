(1) y -b - sin(x - a)의 그래프는 y = sin(x)의 그래프를 x축으로 a, y축으로 b만큼 평행 이동한 것이다.

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.sin(x)

plt.figure(figsize=(12, 5))
plt.plot(x, y, label="sin(x)")
plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi], labels=["-2pi", "-pi", "0", "pi", "2pi"])
plt.grid(linestyle=":")
plt.legend()
plt.show()
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
# y = np.sin(x)
y = np.sin(x - 2)

plt.figure(figsize=(12, 5))
plt.plot(x, y, label="sin(x)")
plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi], labels=["-2pi", "-pi", "0", "pi", "2pi"])
plt.grid(linestyle=":")
plt.legend()
plt.show()
```

(2) y = sin(x - a) + b의 그래프는 y = sin(x)의 그래프를 x축으로 a, y축으로 b만큼 평행 이동한 것이다.

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.sin(x)
y1 = np.sin(x - np.pi)

plt.figure(figsize=(12, 5))
plt.plot(x, y, label="sin(x)")
plt.plot(x, y1, label="sin(x - pi)")
plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi], labels=["-2pi", "-pi", "0", "pi", "2pi"])
plt.grid(linestyle=":")
plt.legend()
plt.show()
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.sin(x)
y1 = np.sin(x - np.pi) + 2

plt.figure(figsize=(12, 5))
plt.plot(x, y, label="sin(x)")
plt.plot(x, y1, label="sin(x - pi) + 2")
plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi], labels=["-2pi", "-pi", "0", "pi", "2pi"])
plt.grid(linestyle=":")
plt.legend()
plt.show()
```

(3) y = -sin(x)의 그래프는 y=sin(x)의 그래프를 x축을 중심으로 선대칭이다.

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.sin(x)
y1 = -np.sin(x)

plt.figure(figsize=(12, 5))
plt.plot(x, y, label="sin(x)", linestyle=":")
plt.plot(x, y1, label="-sin(x)")
plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi], labels=["-2pi", "-pi", "0", "pi", "2pi"])
plt.grid(linestyle=":")
plt.legend()
plt.show()
```

(4) y = -cos(x)의 그래프는 y=cos(x)의 그래프를 x축을 중심으로 선대칭이다.

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.cos(x)
y1 = -np.cos(x)

plt.figure(figsize=(12, 5))
plt.plot(x, y, label="cos(x)", linestyle=":")
plt.plot(x, y1, label="-cos(x)")
plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi], labels=["-2pi", "-pi", "0", "pi", "2pi"])
plt.grid(linestyle=":")
plt.legend()
plt.show()
```

(5) y = a*sin(x)의 그래프는 y = sin(x)의 함숫값을 a배한 것이다.

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.sin(x)
y1 = 2*np.sin(x)

plt.figure(figsize=(12, 5))
plt.plot(x, y, label="sin(x)", linestyle=":")
plt.plot(x, y1, label="2*sin(x)")
plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi], labels=["-2pi", "-pi", "0", "pi", "2pi"])
plt.grid(linestyle=":")
plt.legend()
plt.show()
```

(6) y = cos(a*x)의 그래프는 y = cos(x)와 모양은 같고 주기가 2pi/|a|인 그래프이다.


```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.cos(x)
y1 = np.cos(2*x)

plt.figure(figsize=(12, 5))
plt.plot(x, y, label="cos(x)", linestyle=":")
plt.plot(x, y1, label="cos(2x)")
plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi], labels=["-2pi", "-pi", "0", "pi", "2pi"])
plt.grid(linestyle=":")
plt.legend()
plt.show()
```

(7) y=sin(s-(pi/2)와 y=sin(s-(pi/2)+1))

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.sin(x-(np.pi/2))
y1 = np.sin(x-(np.pi/2))+1

plt.figure(figsize=(12, 5))
plt.plot(x, y, label="sin(x)", linestyle=":")
plt.plot(x, y1, label="sin(2x)")
plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi], labels=["-2pi", "-pi", "0", "pi", "2pi"])
plt.grid(linestyle=":")
plt.legend()
plt.show()
```

감쇠 진동 함수

```python
# 변수 정의
A = 1  # 진폭
omega = 2 * np.pi  # 각진동수 (주기 1초)
alpha = 0.1  # 감쇠 계수
t = np.linspace(0, 20, 1000)  # 시간

# 감쇠 진동 함수 정의
y = A * np.cos(omega * t) * np.exp(-alpha * t)

# 그래프 그리기
plt.figure(figsize=(10, 6))
plt.plot(t, y, label=r'$y(t) = A \cdot \cos(\omega t) \cdot e^{-\alpha t}$')
plt.title('Damped Oscillation')
plt.xlabel('Time (t)')
plt.ylabel('Displacement (y)')
plt.grid(True)
plt.legend()
plt.show()
```

하트 그래프

```python
# t 값을 생성 (0부터 2π까지 0.01 간격으로)
t = np.linspace(0, 2 * np.pi, 1000)

# x와 y 좌표 계산
x = 16 * np.sin(t)**3
y = 13 * np.cos(t) - 5 * np.cos(2*t) - 2 * np.cos(3*t) - np.cos(4*t)

# 그래프 그리기
plt.figure(figsize=(6, 6))
plt.plot(x, y, color='red')

# 축 비율을 동일하게 설정
plt.axis('equal')

# 제목 추가
plt.title('Heart Shape')

# 그래프 출력
plt.show()
```

꽃모양 그래프

```python
# θ 값을 생성 (0부터 4π까지 0.01 간격으로)
theta = np.linspace(0, 4 * np.pi, 1000)

# 꽃 모양 좌표 계산
r = np.sin(6 * theta)  # k=6, 6개의 꽃잎
x = r * np.cos(theta)
y = r * np.sin(theta)

# 그래프 그리기
plt.figure(figsize=(6, 6))
plt.plot(x, y, color='magenta')

# 축 비율을 동일하게 설정
plt.axis('equal')

# 제목 추가
plt.title('Flower Shape')

# 그래프 출력
plt.show()
```

나선형 그래프

```python
# θ 값을 생성 (0부터 4π까지 0.01 간격으로)
theta = np.linspace(0, 4 * np.pi, 1000)

# 나선형 좌표 계산
a = 0
b = 0.1
r = a + b * theta
x = r * np.cos(theta)
y = r * np.sin(theta)

# 그래프 그리기
plt.figure(figsize=(6, 6))
plt.plot(x, y, color='blue')

# 축 비율을 동일하게 설정
plt.axis('equal')

# 제목 추가
plt.title('Spiral Shape')

# 그래프 출력
plt.show()
```

그래디언트 원형 그래프

```python
# θ 값을 생성 (0부터 2π까지 0.01 간격으로)
theta = np.linspace(0, 2 * np.pi, 1000)

# 그래디언트 원형 좌표 계산
r = np.ones_like(theta)
x = r * np.cos(theta)
y = r * np.sin(theta)

# 그래프 그리기
plt.figure(figsize=(6, 6))
plt.scatter(x, y, c=theta, cmap='plasma', s=1)  # 색상 변화

# 축 비율을 동일하게 설정
plt.axis('equal')

# 제목 추가
plt.title('Gradient Circular Pattern')

# 그래프 출력
plt.show()
```

페르세폴리스

```python
# θ 값을 생성 (0부터 2π까지 0.01 간격으로)
theta = np.linspace(0, 2 * np.pi, 1000)

# 별 모양 좌표 계산
n = 5  # 별의 꼭짓점 수
r = np.sin(n * theta)
x = r * np.cos(theta)
y = r * np.sin(theta)

# 그래프 그리기
plt.figure(figsize=(6, 6))
plt.plot(x, y, color='green')

# 축 비율을 동일하게 설정
plt.axis('equal')

# 제목 추가
plt.title('Pentagon Star Shape')

# 그래프 출력
plt.show()
```

폴라 로즈 (Polar Rose)

```python
# θ 값을 생성 (0부터 2π까지 0.01 간격으로)
theta = np.linspace(0, 2 * np.pi, 1000)

# 폴라 로즈 좌표 계산
k = 6  # 곱하기 값
r = 1 + 0.5 * np.sin(k * theta)
x = r * np.cos(theta)
y = r * np.sin(theta)

# 그래프 그리기
plt.figure(figsize=(6, 6))
plt.plot(x, y, color='green')

# 축 비율을 동일하게 설정
plt.axis('equal')

# 제목 추가
plt.title('Polar Rose with k=6')

# 그래프 출력
plt.show()
```
