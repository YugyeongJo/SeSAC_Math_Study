```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 예제 3.3.8 실습

```python
# Define the functions
def f1(x):
    return 2**(x - 2) + 1

def f2(x):
    return 2**(-x + 2) + 1

# Generate x values
x = np.linspace(-5, 5, 400)

# Create a pandas DataFrame for the x values and corresponding y values
df = pd.DataFrame({
    'x': x,
    'y1': f1(x),
    'y2': f2(x)
})

# Plotting using pandas
plt.figure(figsize=(8, 6))
df.plot(x='x', y='y1', label=r'$y = 2^{x - 2} + 1$', color='blue', ax=plt.gca())
df.plot(x='x', y='y2', label=r'$y = 2^{-x + 2} + 1$', color='red', ax=plt.gca())

# Adding asymptote
plt.axhline(y=1, color='gray', linestyle='--', label='Asymptote $y=1$')

# Adding titles and labels
plt.title('Graphs of $y = 2^{x - 2} + 1$ and $y = 2^{-x + 2} + 1$ (using pandas)')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1, 5)
plt.xlim(-5, 5)

# Show legend
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

## 예제 3.3.11

```python
# Define the x values
x = np.linspace(1.1, 10, 400)  # x values from just above 1 to 10

# Define the functions
y1 = np.log2(x - 1) + 2
y2 = np.log2(x - 1) / np.log2(1/2) + 2  # Using change of base formula for log(1/2)

# Create the plot
plt.figure(figsize=(10, 6))

# Plot y = log2(x-1) + 2
plt.plot(x, y1, label='$y = \log_2(x - 1) + 2$', color='blue')

# Plot y = log(1/2)(x-1) + 2
plt.plot(x, y2, label='$y = \log_{1/2}(x - 1) + 2$', color='red', linestyle='--')

# Add labels and title
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Graphs of $y = \log_2(x - 1) + 2$ and $y = \log_{1/2}(x - 1) + 2$')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()

# Show the plot
plt.show()
```

## y = log2(x-1)+2 와 y = log(1/2)(x-1)+2의 그래프를 그리고 점근선, 정의역을 구하라

```python
x = np.linspace(0, 4, 100)

y1 = np.log2(x-1)+2
y2 = np.log2(x-1)/np.log2(1/2)+2

plt.figure(figsize=(10, 6))

plt.plot(x, y1, label='$y = \log_2(x - 1) + 2$', color='blue')
plt.plot(x, y2, label='$y = \log_{1/2}(x - 1) + 2$', color='red', linestyle='--')

plt.axvline(x=1, color='black', linestyle=':', label='Vertical Asymptote $x = 1$')

plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Graphs of $y = \log_2(x - 1) + 2$ and $y = \log_{1/2}(x - 1) + 2$')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()

plt.show()
```

## Binary Cross Entropy
1) y = -logx
2) y2 = -log(1-x)

```python
# Binary loss function

x = np.linspace(0, 4, 100)
y1 = -np.log(x)
y2 = -np.log(1-x)

plt.plot(x, y1, label="-log(x)")
plt.plot(x, y2, label="-log(1-x)")

plt.grid(linestyle = ":")
plt.xlim(0, 1)
plt.vlines(1, -2, 3, colors="violet", linestyles="-.")
plt.legend()
plt.show()
```

## 하트그래프 그리기

```python
# Define the parameter t
t = np.linspace(0, 2 * np.pi, 1000)

# Define the heart shape using parametric equations
x = 16 * (np.sin(t))**3
y = 13 * np.cos(t) - 5 * np.cos(2 * t) - 2 * np.cos(3 * t) - np.cos(4 * t)

# Plot the heart shape
plt.figure(figsize=(8, 8))
plt.plot(x, y, color='red')

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Heart Shape')

# Add a grid
plt.grid(True)

# Set equal aspect ratio
plt.gca().set_aspect('equal', adjustable='box')

# Show the plot
plt.show()
```
