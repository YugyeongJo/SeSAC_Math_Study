```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v = np.array([1, 1])
w = np.array([-2, -1])

plt.quiver(0, 0, v[0], v[1], angles = "xy", scale_units = "xy", scale = 1, color = "r")
plt.quiver(0, 0, w[0], w[1], angles = "xy", scale_units = "xy", scale = 1, color = "b")

plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.grid(linestyle=":")
plt.show()
```
